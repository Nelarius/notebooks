```python
import matplotlib.pyplot as ply
import numpy as np
```

```python
def point_in_sphere(t, u, v):
    r = np.cbrt(t)
    cos_theta = 1 - 2.0 * u
    sin_theta = np.sqrt(1 - cos_theta * cos_theta)
    phi = 2 * np.pi * v

    x = r * sin_theta * np.cos(phi)
    y = r * sin_theta * np.sin(phi)
    z = r * cos_theta

    return x, y, z
```

```python
r = np.random.uniform(0, 1, 1000)
u = np.random.uniform(0, 1, 1000)
v = np.random.uniform(0, 1, 1000)
```

```python
points = [point_in_sphere(r[i], u[i], v[i]) for i in range(1000)]
x, y, z = zip(*points)

fig = ply.figure()
ax = ply.axes(projection='3d')
ax.scatter3D(x, y, z, c=z, cmap='viridis')
ply.show()
```

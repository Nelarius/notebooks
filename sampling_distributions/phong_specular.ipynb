# Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Definitions

We want to draw samples out of the distribution

$$
f(\omega_o) = \frac{n+1}{2\pi}\cos^n(\alpha),
$$

where $\alpha$ is the angle between the reflection of $\omega_i$ and $\omega_o$, i.e. $\mathbf{R} \cdot \mathbf{L}$. The direction can be specified in terms of $(\alpha, \, \phi)$, where:

$$
(\alpha, \, \phi) = (\arccos(u^{\frac{1}{n+1}}_1), \, 2 \pi u_2)
$$

In cartesian coordinates, this becomes

$$
\begin{aligned}
x &= \sqrt{1 - u^{\frac{2}{n+1}}_1}\cos (2\pi u_2), \\
y &= \sqrt{1 - u^{\frac{2}{n+1}}_1} \sin (2 \pi u_2), \\
z &= u^{\frac{1}{n+1}}_1
\end{aligned}
$$

```python
def point_in_specular_distribution(u1, u2, n):
    x = np.sqrt(1 - u1**(2/(n+1))) * np.cos(2*np.pi*u2)
    y = np.sqrt(1 - u1**(2/(n+1))) * np.sin(2*np.pi*u2)
    z = u1**(1/(n+1))

    return x, y, z
```

# Random points in a specular distribution

The points are distributed about the z-axis, and need to be transformed into world-space.

```python
n = 10.0
u1 = np.random.uniform(0, 1, 500)
u2 = np.random.uniform(0, 1, 500)
```

```python
points = [point_in_specular_distribution(u1, u2, n) for u1, u2 in zip(u1, u2)]
x, y, z = zip(*points)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(x, y, z, c=z, cmap='viridis', linewidth=0.5)
ax.set_title('Specular sample distribution')
plt.show()
```

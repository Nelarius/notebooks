# Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Definitions

```python
def tri_point_at_parameter(tri, r1, r2):
    # https://math.stackexchange.com/questions/18686/uniform-random-point-in-triangle-in-3d
    return (1 - np.sqrt(r1)) * tri[0] + (np.sqrt(r1) * (1 - r2)) * tri[1] + (np.sqrt(r1) * r2) * tri[2]
```

# Random points in a triangle

```python
num_points = 500
u = np.random.uniform(0.0, 1.0, num_points)
v = np.random.uniform(0.0, 1.0, num_points)
```

```python
tri = (np.array([-1, 0, 0]), np.array([1, 0, 0]), np.array([0, 1, 0]))
points = [tri_point_at_parameter(tri, u, v) for u, v in zip(u, v)]
x, y, z = zip(*points)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(x, y, z, c=x, cmap='Reds')
ax.set_title("Uniform triangle distribution")
plt.show()
```

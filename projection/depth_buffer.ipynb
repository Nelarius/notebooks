```python
# Imports

import numpy as np
import matplotlib.pyplot as plt
```

```python
def nonlinear_depth(ze, n, f):
    return (f + n) / (f - n) + 2.0 * f * n / ((f - n) * ze)
```

```python
# Plot nonlinear_depth in the range 1, 2005
zs = np.linspace(-2005.0, -1.0, 500)

plt.title("n = 1, f = 2005")
plt.plot(zs, nonlinear_depth(zs, 1.0, 2005.0))
plt.xlabel("z_e")
plt.ylabel("z_ndc")
```

```python
zs = np.linspace(-100.0, -2005.0, 100)

plt.title("n = 100, f = 2005")
plt.plot(zs, nonlinear_depth(zs, 100.0, 2005.0))
plt.xlabel("z_e")
plt.ylabel("z_ndc")
```

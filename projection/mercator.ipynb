```python
# Imports

import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
earth_radius_meters = 6378137.0
earth_circumference_meters = 2 * math.pi * earth_radius_meters

def mercator_scale(lat):
    return np.cos(lat * math.pi / 180.0)

def meter_in_mercator_coordinate_units(lat):
    return 1 / earth_circumference_meters * mercator_scale(lat)
```

```python
# Plot meter_in_mercator_coordinate_units as a function of latitudes in the range [-85, 85] degrees.

# Range of latitudes from -85 to 85 degrees
latitudes = np.linspace(-85, 85, 100)

# Plot latitudes vs meter_in_mercator_coordinate_units
plt.title('Latitudes vs meter_in_mercator_coordinate_units')
plt.plot(latitudes, meter_in_mercator_coordinate_units(latitudes))
plt.xlabel('latitude (degrees)')
plt.ylabel('meter_in_mercator_coordinate_units')

```
